# 1. 基本图表绘制
## 1.1 折线图

`plt.plot()`绘制折线图。

````python
import matplotlib.pyplot as plt
x = [1,2,3]
y = [2,4,6]
# 绘制折线图
plt.plot(x,y)
# 展示图形
plt.show()
````

如果让x和y之间存在数学的联系，需要引入`numpy`

````python
#
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([1,2,3])
y1 = x1 *2 +1
plt.plot(x1, y1)
y2=x1**2
plt.plot(x1, y2, color = 'red', linewidth=3, linestyle='--')
plt.show()
````

## 1.2 柱状图
`plt.bar()`绘制柱状图

````python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [5,4,3,2,1]
plt.bar(x,y)
plt.show()
````

## 1.3 散点图

````python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(10)
y = np.random.rand(10)
plt.scatter(x,y)
plt.show
````

1.4 直方图


````python
import matplotlib.pyplot as plt
import numpy as np
# 随机生成10000个服从标准正太分布的数据
data = np.random.randn(10000)
# bins为颗粒度，即直方图的柱形数量； edgecolor 为柱形边框的颜色
plt.hist(data, bins=40, edgecolor = 'black')
plt.show()
````

## 2.1 添加文字说明



````python
import matplotlib.pyplot as plt

# 设置字体
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
x = [-2,-1,3,4]
y=[2,4,6,8]
plt.plot(x, y)
plt.title('标题')
plt.xlabel('x轴')
plt.ylabel('y轴')
plt.show()
````

## 2.2 添加图例



````python
import numpy as np
import matplotlib.pyplot as plt

# 第1条线： 设置label(标签)为 'y = x+1'
x1 = np.array([1,2,3])
y1 = x1 + 1
plt.plot(x1, y1, label = 'y=x+1')

# 第2条线：设置label(标签)为'y=x*2'
y2 = x1*2
plt.plot(x1, y2, color='red', linestyle='--', label='y=x*2')
# 设置图例位置为左上角
plt.legend(loc = 'upper left')
plt.show()
````

## 2.3 设置双坐标



````python
import numpy as np
import matplotlib.pyplot as plt


# 第1条线: 设置label(标签) 为"y=x"
x1 = np.array([10, 20, 30])
y1 = x1
plt.plot(x1, y1, color='red', linestyle='--', label = 'y=x')
plt.legend(loc='upper left')

# 设置双坐标轴
plt.twinx()

# 第2条线: 设置label(标签) 为'y=2^2'
y2 = x1*x1
plt.plot(x1, y2, label = 'y=x^2')
plt.legend(loc = 'upper right')


plt.show()
````

````python
import tushare as ts
import matplotlib.pyplot as plt
import mpl_finance as mpf
import seaborn as sns
sns.set()
````

````python
# 获取股票数据
df = ts.get_k_data('000002', '2019-06-01','2019-09-30')
df.head()
````

````python
# 日期格式的转换
from matplotlib.pylab import  date2num
import datetime

# 将tushare库获取的日期数据格式转换
def date_to_num(dates):
    num_time = []
    for date in dates:
        date_time = datetime.datetime.strptime(date, '%Y-%m-%d')
        num_date = date2num(date_time)
        num_time.append(num_date)
    return num_time

df_arr = df.values
df_arr[:,0] = date_to_num(df_arr[:,0])
````

````python
print(df_arr[0:5])
````

````python
# 绘制K线图
fig,ax = plt.subplots(figsize=(15,6))
mpf.candlestick_ochl(ax, df_arr, width=0.6,colorup='r', colordown='g', alpha=1.0)
# 绘制网格线
plt.grid(True)
# 设置X轴的刻度格式为常规日期格式
ax.xaxis_date()
````

参数解释：

| 参数      | 含义                           |
| --------- | ------------------------------ |
| ax        | 绘制Axes的实例，即画布中的子图 |
| df_arr    | 股价历史数据                   |
| width     | K线柱形的宽度                  |
| colorup   | 收盘价高于开盘价的颜色         |
| colordown | 收盘价低于开盘价的颜色         |
| alpha     | 柱形颜色的透明度               |



下面完成一个完整的K线图

* 添加均线图
* 日成交量




````python
import tushare as ts
import matplotlib.pyplot as plt
import mpl_finance as mpf
from matplotlib.pylab import  date2num
import datetime
import seaborn as sns
sns.set()

# mac系统设置字体为'Arial Unicode MS', 即可解决乱码问题
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

df = ts.get_k_data('000002', '2019-06-01', '2019-09-30')

def date_to_num(dates):
    num_time = []
    for date in dates:
        date_time = datetime.datetime.strptime(date, '%Y-%m-%d')
        num_date = date2num(date_time)
        num_time.append(num_date)
    return num_time



# 添加5日均线
df['MA5'] = df['close'].rolling(5).mean()
# 添加10日均线
df['MA10'] = df['close'].rolling(10).mean()


df_arr = df.values
df_arr[:,0] = date_to_num(df_arr[:,0])

fig,axes = plt.subplots(2,1, sharex=True, figsize=(15,6))
ax1, ax2 = axes.flatten()
# 绘制第1张子图：K线图和均线图
mpf.candlestick_ochl(ax1, df_arr, width=0.6,colorup='r', colordown='g',alpha=1.0)
# 5日均线
ax1.plot(df_arr[:,0], df['MA5'])
# 10日均线
ax1.plot(df_arr[:,0], df['MA10'])


# 绘制网格
ax1.grid(True)
# 设置图标标题
ax1.set_title('万科A')
# 设置y轴坐标
ax1.set_ylabel('价格')

ax.xaxis_date()

# 绘制第2张子图：每日成交量柱形图
ax2.bar(df_arr[:,0], df_arr[:,5])
ax2.set_xlabel('日期')
ax2.set_ylabel('成交量')
ax2.grid(True)
ax2.xaxis_date()
````
